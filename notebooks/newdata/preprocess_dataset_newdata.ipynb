```python
import pandas as pd
```

```python
# Load the data
df = pd.read_csv('../../data/data_from_2010_2.csv', engine='python')
```

```python
# Display the data
df.head()
```

```python
# turn EventDate column from Integer to Date
df['EventDate'] = pd.to_datetime(df['EventDate'], format='%Y%m%d')
```

```python
df.head()
```

### Drop data before specified date and drop rows without two actors

```python
drop_before = '2015-01-01' # Change this date to drop data before that date
df = df.drop(df[df.EventDate < drop_before ].index)
df = df.reset_index(drop=True)
df.head()
```

```python
df = df.dropna(subset=['Actor1CountryCode', 'Actor2CountryCode'], ignore_index=True)
df.head()
```

```python
group_by_month = pd.Grouper(key='EventDate', freq='ME')
group_by_actor1_countrycode = pd.Grouper(key='Actor1CountryCode')
group_by_actor2_country_code = pd.Grouper(key='Actor2CountryCode')
```

```python
g_scale_and_avgtone = df.groupby([group_by_month, group_by_actor1_countrycode, group_by_actor2_country_code])[['GoldsteinScale','AvgTone']].mean().reset_index()
g_scale_and_avgtone
```

```python
mentions_and_articles = df.groupby([group_by_month, group_by_actor1_countrycode, group_by_actor2_country_code])[['NumMentions','NumArticles']].mean().reset_index()
mentions_and_articles
```

```python
grouped_data = pd.merge(g_scale_and_avgtone, mentions_and_articles, on=['EventDate', 'Actor1CountryCode', 'Actor2CountryCode'])
grouped_data.columns = ['EventDate', 'Actor1CountryCode', 'Actor2CountryCode', 'GoldsteinScale_Averaged', 'AvgTone_Averaged', 'NumMentions_averaged', 'NumArticles_averaged']
grouped_data
```

```python
grouped_data.to_csv('../../data/processed_data/processed_data_2015_onward_mean_newdata.csv', index=False)
```
