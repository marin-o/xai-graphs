```python
import pandas as pd
import networkx as nx
```

```python
# Load the data
df = pd.read_csv('data/data_from_2010.csv', low_memory=False)
```

```python
# Display the data
df.head()
```

```python
# turn EventDate column from Integer to Date
df['EventDate'] = pd.to_datetime(df['EventDate'], format='%Y%m%d')
```

```python
df.head()
```

```python
df['NewEventDate'] = df['EventDate'].dt.strftime('%Y-%m')
```

```python
df.head()
```

```python
grouper1 = pd.Grouper(key='EventDate', freq='ME')
grouper2 = pd.Grouper(key='Actor1CountryCode')
grouper3 = pd.Grouper(key='Actor2CountryCode')
```

```python
df.groupby([grouper1, grouper2, grouper3])['GoldsteinScale'].sum()
```

```python
avg_g_scale = df.groupby(['NewEventDate','Actor1CountryCode', 'Actor2CountryCode'])['GoldsteinScale'].sum()
```
